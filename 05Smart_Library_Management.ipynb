## Project Explanation
This project aims to show a library management system with OOP. There are "member" and "book" objects which have their own common attributes. (Such as name for member, title for book). It is easier to collect these attributes in one class and use them for each object. 
Python Topics: 
OOP (classes, methods, inheritance)


```python
class Book():
    def __init__(self, title, author, year, is_available):
        self.title = title
        self.author = author
        self.year = year
        self.is_available = True

    def get_info(self):
        print(f"Book information -> Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year: {self.year}")
        print(f"Is it available? : {self.is_available}")
```

```python
class Member():
    def __init__(self, name, member_id):
        #Book.__init__(self)
        self.name = name
        self.member_id = member_id

    
    def borrow_book(self, book):
        self.book = book
        if book.is_available:
            book.is_available = False
            print(f"Book '{book.title}' was taken by '{self.name}'")
        else:
            print(f"Sorry, book '{book.title}' is not available.")

    def return_book(self, book):
        self.book = book
        if book.is_available == False:
            book.is_available = True
            print(f"Book '{book.title}' was returned by '{self.name}'")
        else:
            print(f"Book is already available in library.")
```

```python
##default is necessary to use, a book can not disappear, it is either borrowed or returned. 
```

```python
class DigitalBook(Book):
    def __init__(self, title, author, year, is_available, file_size):
        super().__init__(title, author, year, True)
        self.file_size = file_size

    def download(self):
        print(f"Downloading... File size: {self.file_size} MB")
```
