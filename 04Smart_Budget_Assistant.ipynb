## Project Explanation
This project aims to collect user expenses from user and calculate some results, e.g total expenses, average of the expenses, min and max expenses, to make an analysis about budget plans. 
Python Topics: 
Functions,
For loops,
Built-in functions: sum(),len(), max(), min()
*args 
try-except 


## get_expenses Function

```python
expenseNum = int(input("How many expenses will you enter?"))
expenses = []
def get_expenses():
    try:
        
            
        for i in range(1, expenseNum + 1):
            expense = float(input(f"Enter expense {i}:"))
            expenses.append(expense)
    except:
        print("Please enter a number")

get_expenses()
```

## calculate_summary Function

```python
def calculate_summary(expenses):
    total = 0
    average = 0
    
    for i in expenses:
        total += i

    average = float(round(total / len(expenses)))
    print(f"Total: {total} TL")
    return f"Average: {average} TL"

calculate_summary(expenses)
```

## flexible_total Function

```python
def flexible_total(*args):
    totalSum = int(sum(args))
    print(f"Number of expenses: {len(args)}")
    return (f"Total Sum: {totalSum}")

flexible_total(*expenses)  ## listeyi liste olarak değil açarak gönder: "*"
```

## find_extremes Function

```python
def find_extremes(expensesList):
    maxExpense = max(expensesList)
    minExpense = min(expensesList)
    return (f"Min Expense: {minExpense} TL, Max Expense: {maxExpense} TL")

find_extremes(expenses)
```

## Scope
Reading and printing global variable is possible (try_global_variable and currency)
Changing was not possible. It changed only in one code block: directly after calling.
But the actual value was still TL: print(f"Still TL or Euro? {currency2}").
Global variable can not be changed without the word "global"

```python
##Reading global variable - works
currency = "TL"
def try_global_variable():
    print(f"Global variable? -> {currency}")

try_global_variable()
```

```python
##Changing global variable - error
currency2 = "TL"
def try_global_variable2():
    currency2 = "Euro"
    print(f"Global variable? -> {currency2}")

try_global_variable2()
```
