```python
from utils import get_user_expenses, generate_random_notification
from models import Expense, Budget
from notifications import Notification, EmailNotification, SMSNotification
```

```python
limit = int(input("Enter your budget limit: "))
budget = Budget(limit)

expenses, categories = get_user_expenses()

for i in range(len(expenses)):
    expense = Expense(expenses[i], categories[i])
    print(expense)
    budget.add_expense(expense)

print(f"Your limit: {limit}")
print(f"Your total expense: {budget.total_expense()}")
print(f"Your average expense: {budget.average_expense()}")
budget.is_limit_exceeded()
```

```python
if budget.is_limit_exceeded():    
    notifications = [
        EmailNotification(generate_random_notification()),
        SMSNotification(generate_random_notification())
    ]
    
    for notification in notifications:
        print(notification.send())
else:
    print("Limit aşılmadı.")
```
