## Notification System

This project manages notifications such as Email, SMS, Push by using Python OOP and modular file structure.

### Polymorphism
-----------
It allows different classes to have the same method name, so each class implements that method differently.

It is needed in this project to add different notification types without changing main code.  
It avoids repetition by using the method `send()`.

### Python Topics:
- OOP (Polymorphism) -> send()
- Magic Method __str__ 
- try - except
- Libraries (e.g., datetime, random)
- Modular structure

```python
from notifications import EmailNotification, PushNotification, SMSNotification
from utils import create_message, get_time
```

```python
notification1 = (EmailNotification(create_message(), get_time()))
#type(notification1)
notification2 = (PushNotification(create_message(), get_time()))
notification3 = (SMSNotification(create_message(), get_time()))
notificationList = [notification1, notification2, notification3]
```

```python
for notification in notificationList:
    notification.send()
```

```python
for notification in notificationList:
    print(notification)
```

```python
try:
    empty_email = EmailNotification(" ", get_time())
    empty_email.send()
except ValueError as e:
    print(e) 
```
