## Raw Data

```python
raw_data ="  AhMeT  yILMAZ ;  23 ;  1.78 ;  ahmetYILMAZ@GMAIL.Com   "
```

## Cleaning and Splitting Data

```python
cleaned_data = raw_data.strip().split(";")
cleaned_data
```

## Correct Name Format

```python
corrected_name = cleaned_data[0].strip().lower().title()
corrected_name
```

## Correct Age Format

```python
int_age = int(cleaned_data[1])
int_age
```

```python
float_age = float(int_age) + 10
float_age
```

## Correct Height Format

```python
float_height = float(cleaned_data[2]) * 100
float_height
```

## Correct Mail Format

```python
corrected_email = cleaned_data[3].strip().lower()
corrected_email
```

```python
index = corrected_email.find("@")
code_email = corrected_email[0:index][:3]
code_email
```

## FINAL PRINT

```python
print("User: " + corrected_name)
print(f"Age (10 years after): {int_age}")
print(f"Height (cm): {float_height}")
print("Email code: " + code_email)
```
