## Random Survey Data

```python
import random

scores = []

while len(scores) < 20:
    score = random.randint(0,6)
    if score == 0:
        break
    if score == 6:
        continue
    scores.append(score)

##we have while here because the loop needs to run for an unknown number of iterations (but stay until 20, and if 0 comes, finish immediately):
##since we don't know when these conditions will happen
```

## Printing Data

```python
scores
```

## Valid Data  (For Loop)

```python
counterValid = 0
counterInvalid = 0

for element in scores:
    if element == 0 or element == 6:
        counterInvalid += 1
    else:
        counterValid += 1

print(f"Valid surveys: {counterValid}")
```

## Average

```python
average = 0
total = 0

for element in scores:
    if element == 0 or element == 6:
        pass
    else:
        total += element

average = total / len(scores)
rAverage = round(average, 2)

print(f"Average: {average}")
print(f"Rounded average: {rAverage}")
```

## Max and Min

```python
minValue = min(scores)
maxValue = max(scores)
print(f"Min value {minValue}, Max Value {maxValue}")
```

## Satisfaction Analysis

```python
satisfiedCounter = 0
dissatisfiedCounter = 0
neutralCounter = 0
oneCounter = 0

for i in scores:
    if i == 4 or i == 5:   ##if i in (4, 5)
        satisfiedCounter += 1
    elif i == 3:     ##elif i in (3)
        neutralCounter += 1 
    elif i == 2 or i == 1:   ##if i in (1, 2)
        dissatisfiedCounter += 1
        if i == 1:
            oneCounter += 1
            if oneCounter == 3:
                print("Critical dissatisfaction detected!!!")
        else: 
            oneCounter = 0
    else:
        print("Something is wrong")

print(f"Satisfied {satisfiedCounter}, Neutral {neutralCounter}, Dissatisfied {dissatisfiedCounter}")
```

## Last Result

```python
print(f"Total Valid Surveys: {counterValid}")
print(f"Average Point: {rAverage}")
print(f"Minimum Point: {minValue}")
print(f"Maximum Point: {maxValue}")
print(f"Satisfied: {satisfiedCounter}")
print(f"Neutral: {neutralCounter}")
print(f"Dissatisfied: {dissatisfiedCounter}")
```
