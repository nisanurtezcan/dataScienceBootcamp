## RAW ORDERS LIST

```python
orders = [
    {
"customer":"Ahmet",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Zeynep",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]
```

## Order Num Analysis

```python
totalOrders = len(orders)
totalOrders
```

```python
order_paid = 0
order_unpaid = 0

if orders[0]["is_paid"] == True:
    order_paid += 1
else:
    order_unpaid += 1
if orders[1]["is_paid"] == True:
    order_paid += 1
else:
    order_unpaid += 1
if orders[2]["is_paid"] == True:
    order_paid += 1
else:
    order_unpaid += 1
if orders[3]["is_paid"] == True:
    order_paid += 1
else:
    order_unpaid += 1


print(f"Paid: {order_paid}, Unpaid {order_unpaid}")
```

## Customer Name Analysis

```python
myNameList = [orders[0]["customer"], orders[1]["customer"], orders[2]["customer"], orders[3]["customer"]]
myNameList
```

```python
myNameSet = set(myNameList)
myNameSet
```

## Product Analysis

```python
collectProducts = [orders[0]["items"], orders[1]["items"], orders[2]["items"], orders[3]["items"]]
collectProducts
```

```python
fullProductList = orders[0]["items"] + orders[1]["items"] + orders[2]["items"] + orders[3]["items"]
fullProductList
```

```python
uniqueProductSet = set(fullProductList)
uniqueProductSet
```

## Order Status

```python
paidText = "Order confirmed."
unpaidText = "Pending"

if orders[0]["is_paid"] == True:
    print("First order:" , paidText)
else:
    print("Second order:" , unpaidText)
if orders[1]["is_paid"] == True:
    print("Second order:" , paidText)
else:
    print("First order:" , unpaidText)
if orders[2]["is_paid"] == True:
    print("Third order:" , paidText)
else:
    print("Third order:" , unpaidText)
if orders[3]["is_paid"] == True:
    print("Fourt order:" , paidText)
else:
    print("Fourt order:" , unpaidText)
```

```python
secureSummary3 = (orders[2]["customer"], orders[2]["total"])
secureSummary3
```

## Final Result

```python
print(f"Total orders: {totalOrders}")
print(f"Paid order: {order_paid}")
print(f"Unpaid order: {order_unpaid}")
print("Unique Products:", uniqueProductSet)
print("Order summaries:", summaryList)
```
